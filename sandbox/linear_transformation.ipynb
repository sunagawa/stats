```python
import numpy as np, pandas as pd, seaborn as sns, matplotlib.pyplot as plt
%matplotlib inline
```

```python
def visualize(matrix, xlim=[-5, 5], ylim=[-5, 5]):
    array = np.array(matrix)
    vector_count = len(array[0])
    zero_array = np.zeros(vector_count)
    plt.figure()
    ax = plt.gca()
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.set_aspect('equal', adjustable='box')
    ax.set_xticks(np.arange(1), minor=False)
    ax.set_yticks(np.arange(1), minor=False)
    ax.set_xticks(np.arange(xlim[0], xlim[1]), minor=True)
    ax.set_yticks(np.arange(ylim[0], ylim[1]), minor=True)
    ax.grid(which='minor', color='black',linestyle='--', alpha=0.2)
    ax.grid(which='major', color='black',linestyle='-', alpha=0.8)
    # xx = np.array(matrix[0])[0]
    # yy = np.array(matrix[1])[0]
    xx = array[0]
    yy = array[1]
    print(zero_array, xx, yy)
    ax.quiver(zero_array, zero_array, xx, yy, angles='xy', scale_units='xy', scale=1, width=0.01, color='b')
    # x, y = arr[0][0] + arr[0][1], arr[1][0] + arr[1][1]
    # ax.quiver([0], [0], [x], [y], angles='xy', scale_units='xy', scale=1, width=0.01, color='r')
    plt.draw()
    plt.show()
    
def to_matrix(*vectors):
    ihats = []
    jhats = []
    for vector in vectors:
        ihats.append(vector[0])
        jhats.append(vector[1])
    print(ihats, jhats)
    return np.matrix([ihats, jhats])
```
