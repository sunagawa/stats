```python
import pandas as pd, numpy as np, seaborn as sns, matplotlib.pyplot as plt
import math, requests
from functional import seq
from fn import _
%matplotlib inline
```

```python
def average(r):
    return sum(r) / len(r)
def covariance(r1, r2):
    avg1 = average(r1)
    avg2 = average(r2)
    zipped = seq(zip(r1, r2))
    numerator = zipped.map(lambda x: (x[0] - avg1) * (x[1] - avg2)).sum()
    denominator = len(r1) - 1
    return numerator / denominator
def variance(r):
    return covariance(r, r)
def corrcoef(r1, r2):
    return covariance(r1, r2) / math.sqrt(variance(r1) * variance(r2))
```

```python
r1 = [1, 2, 3]
r2 = [3, 4, 4]
corrcoef(r1, r2)
```
